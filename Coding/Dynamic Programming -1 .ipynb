```python
def lcs(X , Y):
    # find the length of the strings
    m = len(X)
    n = len(Y)
 
    # declaring the array for storing the dp values
    L = [[0]*(n+1) for i in range(m+1)]
 
    """Following steps build L[m+1][n+1] in bottom up fashion
    Note: L[i][j] contains length of LCS of X[0..i-1]
    and Y[0..j-1]"""
    for i in range(m+1):
        for j in range(n+1):
            
            for li in L:
                print(li)
            print("--------------------------")
            
            if i == 0 or j == 0 :
                L[i][j] = 0
            
            elif X[i-1] == Y[j-1]:
                L[i][j] = L[i-1][j-1]+1
            
            else:
                L[i][j] = max(L[i-1][j] , L[i][j-1])
                
    for li in L:
        print(li)
        print("--------------------------")
 
    # L[m][n] contains the length of LCS of X[0..n-1] & Y[0..m-1]
    return L[m][n]
#end of function lcs
 
 
# Driver program to test the above function
X = "AGGTAB"
Y = "GXTXAYB"
print (lcs("asd", "qsd"))
 
```

```python
def LCSubStr(X, Y, m, n):
    LCSuff = [[0 for k in range(n+1)] for l in range(m+1)]

    result = 0

    for i in range(m + 1):
        for j in range(n + 1):
            
            for row in LCSuff:
                print(row)
            print("---------------------------")
            
            if (i == 0 or j == 0):
                LCSuff[i][j] = 0
            elif (X[i-1] == Y[j-1]):
                LCSuff[i][j] = LCSuff[i-1][j-1] + 1
                result = max(result, LCSuff[i][j])
            else:
                LCSuff[i][j] = 0
    return result

s = "abcda"
s1  = "abcdssa"
        
LCSubStr(s, s1, len(s), len(s1))  
```

```python
def lcm(s, s1):
    m = len(s)
    n = len(s1)
    result = 0
    matrix = [[0]*(n+1)]*(m+1)
    
    for i in range(1, n):
        for j in range(1, m):
            
            for row in matrix:
                print(row)
            print("-----------------------")
            
            if s[j] == s1[i]:
                matrix[i][j] = matrix[i-1][j-1] + 1
                result = max(result, matrix[i][j])
            
            
    return result
            
         
s = "abcd"
s1  = "abcdssa"
        
lcm(s, s1)   
```
