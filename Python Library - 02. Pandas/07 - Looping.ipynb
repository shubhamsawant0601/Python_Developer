```python
import pandas as pd 
data = pd.read_csv("Automobile_data.csv")
data.head()
```

### Looping through Rows

```python
for index,row in data.iterrows():
    print(index, row["company"], row["length"])
```

```python
for group_key, group in data.groupby(['body-style']):
    print(group_key)
    print('------------------')
    print(group)
    print('==================\n')
```

### Looping through Group

```python
for group_key, group in data.groupby(['body-style', 'company']):
    print(group_key)
    print('------------------')
    print(group)
    print('==================\n')
```

### Looping through Group by with more than two columns

```python
for group_key, group in data.groupby(['body-style', 'company']):
    print(group_key)
    print('------------------')
    print(group)
    print('==================\n')
```
