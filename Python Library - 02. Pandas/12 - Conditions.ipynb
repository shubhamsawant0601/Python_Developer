```python
import pandas as pd
data = pd.read_csv("nba.csv")
data.head()
```

# Conditional Filtering 

```python
# conditional 
filter1 = data["Number"]>5
filter2 = data["Number"]<50
filter3 = data["Age"]>20
filter4 = data["Age"]<30
filter5 = data["Weight"].isnull()==False

data.where(filter1 & filter2 & filter3 & filter4)
```

```python
# Data filtering with where method
data['Indexes'] = data["Name"].str.find('J')==0
data  = data.where(data['Indexes']==True)
data.dropna()
```
