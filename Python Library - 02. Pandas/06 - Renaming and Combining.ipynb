```python
import pandas as pd
data = pd.read_csv("forestfires.csv")
pd.set_option("display.max_rows", 5)
data
```

# Renaming 
1. `rename()` - [columns, index] - columns and index are given Dict of Key-Value Pair, Key as *name to be changed* and value as *changed value*.
2. `rename_axis("NametoChange", axis="para")` - para=["columns",rows]

```python
data.rename(columns={'month':'MONTHS', 'day':'DAYS'})
```

```python
data.rename(index={0:'First', 1:'Second'})
```

```python
data.rename_axis("ROWS", axis='rows').rename_axis("COLUMNS", axis='columns')
```

# Combining
1. `concat()`
2. `join()` - `right.join`, `left.join`para-[lsuffix,rsuffix]
3. `merge()`

### Concat()

```python
x = data.loc[0:10,"X":"Y"]
y = data.loc[0:10,"month":"day"]
pd.concat([x,y])
```

### join()

x.join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False) 

```python
x = data.loc[0:10,"X":"month"]
y = data.loc[0:10,"month":"day"]
```

```python
print(x)
print(y)
```

```python
joined = x.join(y, rsuffix="_right", lsuffix="_left")
joined
```

```python
x = data.loc[0:10,"X":"Y"]
y = data.loc[0:10,"month":"day"]
```

```python
joined = x.join(y)
joined
```
