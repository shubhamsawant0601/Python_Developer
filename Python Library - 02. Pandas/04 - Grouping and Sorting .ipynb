# 1. Merge, Join, Concatenate and Campare

## Concatenating objects 
#### 1. concate()

```python
df1 = pd.DataFrame(
    {
        "A":["A0","A1","A2","A3"],
        "B":["B0","B1","B2","B3"],
        "C":["C0","C1","C2","C3"],
        "D":["D0","D1","D2","D3"],
    },
    index = [1,2,3,4],
)

df2 = pd.DataFrame(
     {
         "A": ["A4", "A5", "A6", "A7"],
         "B": ["B4", "B5", "B6", "B7"],
         "C": ["C4", "C5", "C6", "C7"],
         "D": ["D4", "D5", "D6", "D7"],
     },
     index=[4, 5, 6, 7],
)

df3 = pd.DataFrame(
     {
         "A": ["A8", "A9", "A10", "A11"],
         "B": ["B8", "B9", "B10", "B11"],
         "C": ["C8", "C9", "C10", "C11"],
         "D": ["D8", "D9", "D10", "D11"],
     },
     index=[8, 9, 10, 11],
)

frames = [df1, df2, df3]

result = pd.concat(frames)   

result
```

# 2. Grouping 
1. **Splitting** - the data into groups based on some criteria.

2. **Applying** - a function to each group independently.

3. **Combining** - the results into a data structure.

1. GroupBy.size()
2. GroupBy.count()
3. GroupBy.first(), GroupBy.last()
4. GroupBy.head(), GroupBy.tail()
5. GroupBy.max(), GroupBy.min(), GroupBy.mean()

```python
data
```

```python
data.groupby(["X"]).groups
```

```python
data.groupby("X").agg(["min","max","mean","sum"])
```

```python
data.groupby('day').day.size()
```

```python
data.groupby('day').day.count()
```

```python
data.groupby('day').temp.min()
```

# 3. Multiple Indexes

Use `as_index = False` to remove multindex after groupby
Or simply use `reset_index()`

```python
df = data.groupby(["X", "Y"], as_index=True).sum()
print(df)
```

# 4. Stacking and Unstacking

### Unstacking

```python
data = data.unstack()
data
```
