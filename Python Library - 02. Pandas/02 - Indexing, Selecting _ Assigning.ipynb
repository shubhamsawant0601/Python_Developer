## Indexing 
#### iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So 0:10 will select entries 0,...,9. loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.

### 1. pd.loc() - 
Label based indexing, multiple access methods like -
1. Single scalar lebal
2. List of labels
3. Slice objects
4. Boolean Array

Label based indexing means we use labels to get the range of rows or columns. e.g.- ['a':'b', "column1" : "column2"].

### 2. pd.iloc() - 
integer based indexing, multiple access methods like -
1. Integer
2. List of integers
3. Range of Values

Intger based indexing means we use intgers to get the range of rows or columns. e.g.- [0:5, 1:5].

```python
data.iloc[0]
```

```python
data.loc[0]
```

```python
data.iloc[5,5]
```

```python
data.iloc[1:4,2]
```

```python
data.iloc[1:5,3]
```

```python
data.loc[:5,["X","Y","month"]]
```

```python
data.loc[1:5,"X":"month"]
```

# Conditional Selection
Conditional selection is done by using `.loc` indexing.

```python
data.loc[data.day == 'fri'].head()
```

```python
data.loc[data.day == 'sun'].head()
```

```python
data.loc[((data.day == 'sun')|(data.day == 'fri'))]
```

```python
data.loc[data.temp>18.00]
```

```python
data.loc[data.day.isin(['sun','sat'])].head()
```

```python
data.loc[data.temp.notnull()].head()
```

```python
data.loc[data.temp.isnull()].head()
```
